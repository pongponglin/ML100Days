### 3. 如何利用 list(...) 實現 a.tolist() 的效果？試著用程式實作。


````python
a = np.random.randint(10, size=6) 

print(a.tolist())
print(list(a))
````

````python
b = np.random.randint(10, size=(3,4)) 

print(b.tolist())
print(list(b))
````

````python
c = np.random.randint(10, size=(2,3,2)) 

print(c.tolist())
print(list(c))
````

````python
def tolist(iterable):
    if iterable.ndim ==1:
        return list(iterable)
    elif iterable.ndim ==2:
        return [list(i) for i in list(b)]
    else:
        return [list(list(j) for j in i) for i in list(c)]

print(tolist(a))
print(tolist(b))
print(tolist(c))
````

````python
## Answer
def tolist(iterable):
    if type(iterable) != np.ndarray:
        return iterable
    return [tolist(i) for i in iterable]

print(tolist(a))
print(tolist(b))
print(tolist(c))
````

````python
import numpy as np 

print(np.searchsorted([1,2,4,3,5], 3))
print(np.searchsorted(
    [1, 2, 3, 4, 5],
    [1, 10, 3, 5]
))
````

````python
a = np.random.rand(5)
print(a)
print(a.sum()) 
print(np.sum(a))
print(sum(a)) 
````

````python
a = np.random.rand(3,4)
print(a)
print(a.sum()) 
print(np.sum(a))
print(sum(a)) 
````

````python
a = np.random.rand(2,3,2)
print(a)
print(a.sum()) 
print(np.sum(a))
print(sum(a)) 
print(a.sum(axis=0))
````
